```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = "https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/master/chapter-02-car-price/data.csv"
df = pd.read_csv(data)
df.head()
```

```python
#cleaning
df.columns = df.columns.str.lower().str.replace(' ', "_")  # make all stringcolumn headings lowercase
df.head()
```

```python
df.dtypes #see all datatypes of each column #objects are usually strings
df.dtypes[df.dtypes == 'object'].index # create an index of all columns of object datatype
strings = list(df.dtypes[df.dtypes == 'object'].index) # turns index into a python string list
strings
```

```python
# change all columns to lower case, replace spaces with _
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
